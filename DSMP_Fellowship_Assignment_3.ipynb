## **Question 1: Basic Class Creation**
Create a Python class called Book that has the following attributes and methods
- Attributes: title, author, pages
- Method: get_info() that returns a string with title, author, and number of pages

Write code to create two objects of the Book class and print their information using the get_info() method.

```python
# Solution
# Create class
class Book:

# Define init method with atttributes
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

# Define method of get_info to return value
    def get_info(self):
        return f"Title: {self.title}, Author: {self.author}, Pages: {self.pages}"

# Create objects of the class
book1 = Book("Peer-e-Kamil", "Umera Ahmed", 1080)
book2 = Book("Shahab Nama", "Qudratullah Shahab", 300)

print(book1.get_info())
print(book2.get_info())
```

## **Question 2: Class Inheritance**
Create a class person with the following attributes:
- name, age.

Then create class Student that inherits from Person and has an additional attributes student_id.

Write a method in Student to display the students's details, includng the inherited attributes. Create an object of student and print the details.


```python
# Solution

# Create person class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}\nAge: {self.age}")

# Create student class that inherits from person
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")

# Create an object of student class
student1 = Student("Ahmed", 25, 231)

# Print object of class Stuent
student1.display_info()
```

## **Question 3: Writing and Reading Files**
Write a Python program to:
- Create a text file student_data.txt
- Use a for loop to write the names and scores of 5 students to the file in the format:
  - Name: < name >, Score: < score >
- After writing, read the file and print its contents line by line

```python
# Solution

# Student names and their score list
students = [
    ("Asma", 85),
    ("Laiba", 92),
    ("Wajiha", 78),
    ("Umama", 88),
    ("Sana", 95)
]

# write to the file using for loop
with open("student_data.txt", "w") as student_data:
    for name, score in students:
        student_data.write(f"Name: {name}, Score: {score}\n") # write to file

# Read the file and print its content line by iine
with open("student_data.txt", "r") as file:
    content = file.read()
    print(content)
```

## **Question 4: Managing Student data with OOP and File handling**
Create a class studentRecord that handles student records. The class should:
- Have methods to add student's name an score
- Save the record to a text file records.txt
- Have a methode to read and print all student records from the file.

Create an instance od StudentRecord and use it to add at least three students. Then display all te recordds from the file.

```python
# Solution
# create class Student Record
class StudentRecord:
    def __init__(self, file="records.txt"):
        self.file = file

# Function to add student and record
    def add_student(self, student_name, student_score):
        with open(self.file, "a") as file:
            file.write(f"Name: {student_name}, Score: {student_score}\n")
# Fuction to display records
    def display_records(self):
        print("Student Records:")
        with open(self.file, "r") as file:
            content = file.read()
            print(content)



# Create an instance of StudentRecord
student_record = StudentRecord()

# Add at least three students
student_record.add_student("Erum", 85)
student_record.add_student("Fatima", 90)
student_record.add_student("Umer", 78)

# Display all records from the file
student_record.display_records()
```
